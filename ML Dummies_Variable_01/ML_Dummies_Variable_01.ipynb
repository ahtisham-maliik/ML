```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dum_var = pd.get_dummies(df.town)
dum_var
```

```python
merged = pd.concat([df, dum_var], axis = 'columns')
merged
```

```python
reg = linear_model.LinearRegression()
reg.fit(merged[['area', 'monroe township', 'robinsville']], merged.price)
```

```python
reg.predict(merged[['area', 'monroe township', 'robinsville']])
```

```python
reg.score(merged[['area', 'monroe township', 'robinsville']], merged.price)
```
