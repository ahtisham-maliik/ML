```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price, marker = '+', color = 'red')
```

```python
area_df = df.drop('price', axis = 'columns')
area_df
```

```python
price_df = df.price
price_df
```

```python
reg = linear_model.LinearRegression()
reg.fit(area_df, price_df)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
reg.predict([[5000]])
```

```python
new_area_df = pd.read_csv('areas.csv')
new_area_df.head(5)
```

```python
p = reg.predict(new_area_df)
p
```

```python
new_area_df['price'] = p
new_area_df
```

```python
new_area_df.to_csv('predictions.csv', index = False)
new_area_df
```

```python
%matplotlib inline
plt.xlabel('Area', fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.scatter(new_area_df.area, new_area_df.price, color = 'red', marker = '+')
plt.plot(new_area_df.area, reg.predict(new_area_df[['area']]), color = 'blue')
```
